```python
import requests

def fetch_osrm_time_matrix(coordinates):
    # Define the OSRM API endpoint for the table service
    osrm_url = "http://router.project-osrm.org/table/v1/driving/"

    # Format the coordinates into a string as required by the OSRM API
    coord_str = ';'.join([f"{lon},{lat}" for lat, lon in coordinates])

    # Construct the full URL for the API request
    request_url = f"{osrm_url}{coord_str}"

    # Make the API request
    response = requests.get(request_url)
    
    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()

        # Extract the duration matrix from the API response
        time_matrix = data.get("durations", [])
        return time_matrix
    else:
        print("Failed to fetch the time matrix. HTTP status code:", response.status_code)
        return None
time_matrix = fetch_osrm_time_matrix(coordinates)
```

```python
def nearest_neighbor_with_time_limit(time_matrix, start_index, costco_indices, max_time, service_time, unvisited_indices):
    num_locations = len(unvisited_indices)  # Consider only unvisited locations
    visited = [False] * len(time_matrix)  # Full length to accommodate all locations
    visited[start_index] = True

    route1 = [start_index]
    current_index = start_index
    total_time = 0  # Initialize total time

    costco_visited = False

    while len(route1) < num_locations + 1:  # +1 for the depot
        next_index = None
        shortest_time = float('inf')

        for i in unvisited_indices:  # Iterate through unvisited locations
            if not visited[i] and time_matrix[current_index][i] < shortest_time:
                # Predict total time if next location is added
                predicted_total_time = total_time + time_matrix[current_index][i] + service_time

                # Check if adding this location exceeds max time
                if predicted_total_time <= max_time or (i in costco_indices and not costco_visited):
                    next_index = i
                    shortest_time = time_matrix[current_index][i]

        # Break the loop if no next location is found or if adding next location exceeds max time
        if next_index is None:
            break

        # Update route and timings
        route1.append(next_index)
        visited[next_index] = True
        total_time += shortest_time + service_time
        current_index = next_index

        # Check if Costco is visited
        if next_index in costco_indices:
            costco_visited = True

    # Ensure at least one Costco is visited
    if not costco_visited and costco_indices:
        # Find the closest Costco that hasn't been visited
        closest_costco_index = min(costco_indices, key=lambda x: time_matrix[current_index][x] if not visited[x] else float('inf'))
        if closest_costco_index != float('inf'):  # Check if a Costco is found
            route1.append(closest_costco_index)
            total_time += time_matrix[current_index][closest_costco_index]
            visited[closest_costco_index] = True

    # Return to the depot, if not already there
    if current_index != start_index:
        route1.append(start_index)
        total_time += time_matrix[current_index][start_index]

    return route1, total_time
# Generate a list of all location indices except the depot
all_indices_except_depot = [i for i in range(len(location_order)) if i != depot_index]

# Find indices for Costco within the full list
costco_indices = [i for i, name in enumerate(location_order) if 'Costco' in name]
# Generate route1 using the adjusted nearest_neighbor_with_time_limit function
route1_indices, route1_time = nearest_neighbor_with_time_limit(
    time_matrix=time_matrix, 
    start_index=depot_index, 
    costco_indices=costco_indices, 
    max_time=8*60*60,  # 8 hours in seconds
    service_time=15*60,  # 15 minutes in seconds
    unvisited_indices=all_indices_except_depot  # Consider all locations except the depot
)

# Convert route indices back to location names
route1 = [location_order[i] for i in route1_indices]

print("Route1:", route1)
print(f"Total Route1 Time: {route1_time / 3600:.2f} hours")  # Convert seconds back to hours for display
print(f"{len(route1)} visited.")
```

```python
costco_1014_index = location_order.index('Costco #1014')

all_stores = set(location_order)

visited_stores = set(route1)

# Determine unvisited stores, ensuring Costcos are included
unvisited_stores = all_stores - visited_stores

# Convert unvisited store names back to indices for optimization
unvisited_indices = [location_order.index(store) for store in unvisited_stores]

# Exclude Costco #1014 from Costco indices for route2 if it was included in route1
if 'Costco #1014' in route1:
    costco_unvisited_indices = [i for i in unvisited_indices if 'Costco' in location_order[i] and i != costco_1014_index]
else:
    costco_unvisited_indices = [i for i in unvisited_indices if 'Costco' in location_order[i]]

# Now adjust the 'create_route_from_unvisited' function to use 'costco_unvisited_indices'
def create_route_from_unvisited(time_matrix, location_order, depot_index, unvisited_indices, service_time_per_stop=15*60, max_route_time=9*60*60, previous_routes=[]):
    # Determine if Costco #1014 was visited in any previous route
    costco_1014_visited = any('Costco #1014' in route for route in previous_routes)
    costco_1014_index = location_order.index('Costco #1014')

    # Adjust Costco indices based on whether Costco #1014 was visited
    if costco_1014_visited:
        costco_unvisited_indices = [i for i in unvisited_indices if 'Costco' in location_order[i] and i != costco_1014_index]
    else:
        costco_unvisited_indices = [i for i in unvisited_indices if 'Costco' in location_order[i]]

    # Generate route using Nearest Neighbor heuristic with time limit
    route_indices, total_time = nearest_neighbor_with_time_limit(time_matrix, depot_index, costco_unvisited_indices, max_route_time, service_time_per_stop, unvisited_indices)

    # Convert route indices back to location names
    route_names = [location_order[i] for i in route_indices]

    return route_names, total_time / 3600  # Return total time in hours

route2, route2_time = create_route_from_unvisited(
    time_matrix=time_matrix,
    location_order=location_order,
    depot_index=17,
    unvisited_indices=unvisited_indices,
    service_time_per_stop=15*60,
    max_route_time=8*60*60,
    previous_routes=[route1]  # Provide route1 as context
)
print("Route2:", route2)
print('--------------------------------------------------------------')
print(f"Total Route2 Time: {route2_time:.2f} hours")
print('--------------------------------------------------------------')
print(f"Visits {len(route2)} stores.")
```
